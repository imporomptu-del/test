```python
from glob import glob
from ultralytics import YOLO
import os
import math

image_dir = "/mnt/d/seaqr/images/train/kolomverse/0"
label_dir = image_dir.replace("images", "labels")
output_log = "suspect_images.txt"
chunk_size = 100

model = YOLO("/mnt/d/KOLOMVERSE/KOLOMVERSE/scripts/models/yolov8s_kolomverse/weights/best.pt") 

img_paths = glob(os.path.join(image_dir, "*.jpg"))[:100]

suspects = []
for i in range(math.ceil(len(img_paths) / chunk_size)):
    print(f"Rrocessing {i*chunk_size} to {(i+1)*chunk_size}")
    batch = img_paths[i*chunk_size:(i+1)*chunk_size]
    results = model(batch, batch=16, verbose=False)
    for res in results:
        img_path = res.path
        fname = os.path.basename(img_path)
        label_path = os.path.join(label_dir, os.path.splitext(fname)[0] + ".txt")
        
        n_dets = len(res.boxes)
        has_labels = os.path.exists(label_path) and os.path.getsize(label_path) > 0
        
        if n_dets > 0 and not has_labels:
            suspects.append(f"{fname} -> {n_dets} detections, no labels")
            print(fname)

with open("suspect_images.txt", "w") as f:
    f.write("\n".join(suspects))

print(f"Done! Found {len(suspects)} suspect images.")

```
